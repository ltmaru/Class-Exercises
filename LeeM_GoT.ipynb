```python
import networkx as nx
HoD=nx.Graph()
HoD.add_nodes_from(['Alicent','Rhaenys','Larys','Rhaenyra','Harwin','Criston','Viserys','Otto','Aegon','Hobert','Daemon','Ryam','Harrold','Erryk','Aemond','Arryk','Aemma','Laenor','Vaemond','Rhae','Corlys','Lord Beesbury','Laena','Jacaerys','Lcuerys','Joffrey','Lyonel'])
```

```python
Serve=nx.DiGraph()
Serve.add_edges_from([('Otto','Viserys'),('Lyonel','Viserys'),('Larys','Alicent')], label='served')
ServeLabels=nx.get_edge_attributes(Serve, 'label')
```

```python
Siblings=nx.DiGraph()
Siblings.add_edges_from([('Otto','Hobert'),('Viserys','Daemon'),('Corlys','Vaemond'),('Larys','Harwin'),('Aegon','Aemond')], label='siblings')
SiblingsLabels=nx.get_edge_attributes(Siblings, 'label')
```

```python
Parent=nx.DiGraph()
Parent.add_edges_from([('Otto','Alicent'),('Rhaenyra','Jacaerys'),('Rhaenyra','Lcuerys'),('Rhaenyra','Joffrey')], label='parent')
ParentLabels=nx.get_edge_attributes(Parent, 'label')
```

```python
Guardian=nx.DiGraph()
Guardian.add_edges_from([('Criston','Alicent'),('Ryam','Viserys'),('Harrold','Viserys'),('Erryk','Viserys'),('Arryk','Viserys')], label='guardian')
GuardianLabels=nx.get_edge_attributes(Guardian, 'label')
```

```python
Allies=nx.DiGraph()
Allies.add_edges_from([('Viserys','Vaemond'),('Viserys','Corlys')], label='allies')
AlliesLabels=nx.get_edge_attributes(Allies, 'label')
```

```python
Married=nx.DiGraph()
Married.add_edges_from([('Viserys','Aemma'),('Rhaenyra','Laenor'),('Daemon','Rhae'),('Laena','Daemon'),('Corlys','Rhaenys')], label='married')
MarriedLabels=nx.get_edge_attributes(Married, 'label')
```

```python
Killed=nx.DiGraph()
Killed.add_edges_from([('Daemon','Vaemond'),('Daemon','Rhae'),('Criston','Lord Beesbury')], label='killed')
KilledLabels=nx.get_edge_attributes(Killed, 'label')
```

```python
import matplotlib.pyplot as plt
plt.figure(1,figsize=(12,8))
pos=nx.circular_layout(HoD)
nx.draw(HoD, pos, node_color='grey', node_size=5, with_labels=True)

nx.draw_networkx_edges(Serve, pos, edge_color='purple')
nx.draw_networkx_edge_labels(Serve, pos, edge_labels=ServeLabels, label_pos=0.2,
font_color='purple')

nx.draw_networkx_edges(Siblings, pos, edge_color='green')
nx.draw_networkx_edge_labels(Siblings, pos, edge_labels=SiblingsLabels, label_pos=0.2,
font_color='green')

nx.draw_networkx_edges(Married, pos, edge_color='blue')
nx.draw_networkx_edge_labels(Married, pos, edge_labels=MarriedLabels, label_pos=0.2,
font_color='blue')

nx.draw_networkx_edges(Killed, pos, edge_color='red')
nx.draw_networkx_edge_labels(Killed, pos, edge_labels=KilledLabels, label_pos=0.2,
font_color='red')

nx.draw_networkx_edges(Parent, pos, edge_color='orange')
nx.draw_networkx_edge_labels(Parent, pos, edge_labels=ParentLabels, label_pos=0.2,
font_color='orange')

nx.draw_networkx_edges(Allies, pos, edge_color='pink')
nx.draw_networkx_edge_labels(Allies, pos, edge_labels=AlliesLabels, label_pos=0.2,
font_color='pink')

nx.draw_networkx_edges(Guardian, pos, edge_color='grey')
nx.draw_networkx_edge_labels(Guardian, pos, edge_labels=GuardianLabels, label_pos=0.2,
font_color='grey')

plt.show()
```
